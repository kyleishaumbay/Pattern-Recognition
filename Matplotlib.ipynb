
# Matplotlib

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Problem 1

* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

** plot (x,y) on both axes and call your figure object to show it.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax1.plot(x, y)
ax2.plot(x, y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

** use x,y, and z arrays to recreate the plot below. **

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax1.plot(x, z)
ax2.plot(x, y)
ax2.set_xlim([20, 22]) 
ax2.set_ylim([30, 50]) 
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig = plt.subplots(nrows=1, ncols=2)
```

** plot (x,y) and (x,z) on the axes. **

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, 'b--')
ax2.plot(x, z, 'r')
```

** Resize the plot by adding the figsize() argument in plt.subplots() from your previous code.**

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 2))
ax1.plot(x, y, 'b')
ax2.plot(x, z, 'r--')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```
